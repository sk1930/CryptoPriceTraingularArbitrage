### merge BTC_Indicators.csv and BTCPriceWithoutDuplicates.csv

```python
df_Indicators = pd.read_csv("BTC_Indicators.csv",index_col = 'Unnamed: 0')
df_Indicators
```

```python
df_withoutDuplciates = pd.read_csv("BTCPriceWithoutDuplicates.csv",index_col = 'Unnamed: 0')
df_withoutDuplciates
```

```python
df_withoutDuplciates['windowtimestamp1']
```

```python
import datetime
```

```python
def retDate(x):
    y=datetime.datetime.strptime(x, "%m/%d/%Y").date().strftime("%d%b%Y").lower()
    return y


df_withoutDuplciates['dateFormatted'] = df_withoutDuplciates['windowtimestamp1'].apply(retDate)
```

```python
df_withoutDuplciates['dateFormatted']
```

```python
df_withoutDuplciates
```

```python
df_withoutDuplciates['dateFormatted']
```

```python
df_Indicators['dateFormatted']
```

```python
df_merge_Indicators_withoutDuplciates = pd.merge(df_withoutDuplciates, df_Indicators, how ='left', on ='dateFormatted')
```

```python
df_merge_Indicators_withoutDuplciates
```

```python
df_merge_Indicators_withoutDuplciates.to_csv("BTCPriceMergedIndicatorsAndWithoutDuplicates.csv")
```
